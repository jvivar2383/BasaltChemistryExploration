I would like to take a look at the locations of the data I'm analysing. I get some general concepts from https://jakevdp.github.io/PythonDataScienceHandbook/04.13-geographic-data-with-basemap.html

```python
plt.figure(figsize=(10, 8))
m = Basemap(projection='cyl')

m.drawmapboundary(fill_color='lightskyblue')
m.fillcontinents(color='lawngreen',lake_color='aqua')
m.drawcoastlines()
#m.arcgisimage(server = "https://server.arcgisonline.com/arcgis/services", xpixels = 1500, verbose= True)
m.scatter(df["LONGITUDE"],df["LATITUDE"], latlon=True, norm =df["ANALYZED MATERIAL"], c = "black")
plt.savefig("/Users/jenifervivar/Documents/GitHub/OceanBasaltML/basalts thesis/src/visualization/data_source")
plt.show()
```
